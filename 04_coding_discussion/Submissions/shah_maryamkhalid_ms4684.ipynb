```python
# load stopwords and store as a list
f = open('stop_words.csv')
csv_f = csv.reader(f)

stopwords = []

for row in csv_f:
    stopwords.append(row[0])
    
f.close() 
```

### (b) data manipulation

```python
# define function to tokenize variables
def tokenize(text=None):
    '''
    Removes capitalization and punctuation, and breaks up sentences into individual words.
    
    '''   
    
    text = text.lower()
    text = text.replace('.','')
    text = text.replace('-','')
    text = text.replace('?','')
    text = text.replace('[','')
    text = text.replace(']','')
    text = text.replace('!','')
    text = text.replace('(','')
    text = text.replace(')','')
    text = text.replace(',','')
    text = text.replace('“','')
    text = text.replace('”','')
    text = text.replace('’','')
    text = text.replace('$','')
    text = text.replace('"','')
    text = text.replace('"','')
    text_list = text.split()
    text_list2 = [word for word in text_list if word not in stopwords]
    return text_list2
```

```python
# create function to count the number of times a word occurs in the text 
def convert_text_to_dtm(txt):
    '''
    Converts text into a document term matrix.
    '''
    d = dict()
    for word in tokenize(txt):
        if word in d:
            d[word][0] += 1
        else:
            d[word] = [1]
    return pd.DataFrame(d)
```

```python
# create document term matrices
aljazeera_dtm = convert_text_to_dtm(aljazeera)
bbc_dtm = convert_text_to_dtm(bbc)
breitbart_dtm = convert_text_to_dtm(breitbart)
cnn_dtm = convert_text_to_dtm(cnn)
fox_dtm = convert_text_to_dtm(fox)
```

```python
# append document term matrices
dtm = aljazeera_dtm.append(bbc_dtm)
dtm = dtm.append(breitbart_dtm)
dtm = dtm.append(cnn_dtm)
dtm = dtm.append(fox_dtm)

# Fill in any missing values with 0s
dtm.fillna(0,inplace=True)
```

```python
# view final document term matrix
dtm
```

```python
# create vectors
aljazeera_vector = dtm.iloc[0].values
bbc_vector = dtm.iloc[1].values
breitbart_vector = dtm.iloc[2].values
cnn_vector = dtm.iloc[3].values
fox_vector = dtm.iloc[4].values
```

### (c) linear algebra

```python
# normalize vectors by scaling them by their lengths
aljazeera_norm = 1/np.sqrt(np.dot(aljazeera_vector,aljazeera_vector))*aljazeera_vector
bbc_norm = 1/np.sqrt(np.dot(bbc_vector,bbc_vector))*bbc_vector
breitbart_norm = 1/np.sqrt(np.dot(breitbart_vector,breitbart_vector))*breitbart_vector
cnn_norm = 1/np.sqrt(np.dot(cnn_vector,cnn_vector))*cnn_vector
fox_norm = 1/np.sqrt(np.dot(fox_vector,fox_vector))*fox_vector
```

```python
# calculate dot products to get angles between vectors
cos_aljazeera_bbc = np.dot(aljazeera_norm,bbc_norm)
cos_aljazeera_breitbart = np.dot(aljazeera_norm,breitbart_norm)
cos_aljazeera_cnn = np.dot(aljazeera_norm,cnn_norm)
cos_aljazeera_fox = np.dot(aljazeera_norm,fox_norm)
cos_bbc_breitbart = np.dot(bbc_norm,breitbart_norm)
cos_bbc_cnn = np.dot(bbc_norm,cnn_norm)
cos_bbc_fox = np.dot(bbc_norm,fox_norm)
cos_breitbart_cnn = np.dot(breitbart_norm,cnn_norm)
cos_breitbart_fox = np.dot(breitbart_norm,fox_norm)
cos_cnn_fox = np.dot(cnn_norm,fox_norm)
```

```python
# specifying column names for data frame
col_names = ["aljazeera","bbc","breitbart","cnn","fox"]
```

```python
# creating data frame
dat = pd.DataFrame(columns=col_names,index=col_names)
```

```python
# adding calculated values to data frame 
dat.iloc[0,0] = 1
dat.iloc[0,1] = cos_aljazeera_bbc
dat.iloc[0,2] = cos_aljazeera_breitbart
dat.iloc[0,3] = cos_aljazeera_cnn
dat.iloc[0,4] = cos_aljazeera_fox
dat.iloc[1,0] = cos_aljazeera_bbc
dat.iloc[1,1] = 1
dat.iloc[1,2] = cos_bbc_breitbart
dat.iloc[1,3] = cos_bbc_cnn
dat.iloc[1,4] = cos_bbc_fox
dat.iloc[2,0] = cos_aljazeera_breitbart
dat.iloc[2,1] = cos_bbc_breitbart
dat.iloc[2,2] = 1
dat.iloc[2,3] = cos_breitbart_cnn
dat.iloc[2,4] = cos_breitbart_fox
dat.iloc[3,0] = cos_aljazeera_cnn
dat.iloc[3,1] = cos_bbc_cnn
dat.iloc[3,2] = cos_breitbart_cnn
dat.iloc[3,3] = 1
dat.iloc[3,4] = cos_cnn_fox
dat.iloc[4,0] = cos_aljazeera_fox
dat.iloc[4,1] = cos_bbc_fox
dat.iloc[4,2] = cos_breitbart_fox
dat.iloc[4,3] = cos_cnn_fox
dat.iloc[4,4] = 1
```
