```python
import csv
with open('us-states.csv') as file:
        state_covid_data = []
        for row in csv.reader(file):
            state_covid_data.append(row)

##len(state_covid_data)
```

```python
##Part 1: How many unique dates are there in this data set?
def unique_dates(lst):
    """Finds # of unique dates in a nested list where 'date' is the first item in each sublist"""
    
    """This function takes a nested list, creates a new list consisting of the first item in each sublist,
    then converts that new list to a set, finds the length of the set, and prints that value."""
    new_list = list(item [0] for item in lst) ##Create a new list of all the items in position 0 in each sublist
    new_set = set(new_list) ##Convert that new list into a set
    unique_count = len(new_set) ##Find the length of that new set. NOTE: The set function will not include the 'data'
                                ##item from the first sublist, so there is no need to add a -1 to get the right unique count
    print(unique_count) ##Print that value
    
unique_dates(state_covid_data) ##Run the function! Spoiler alert: There are 226 unique dates in this data set.
```

```python
##Part 2: Find the first date in which the District of Columbia recorded a case.
def firstDC(lst):
    """Finds earliest date when the District of Columbia recorded a case of COVID-19"""
    
    """This function takes each sublist, tests whether or not the observation in position [1] is equal to District of
    Columbia, and if so, appends the observation in position [0] to a new list. It then prints the minimum value in
    that new list, which is the earliest date the district recorded a case."""
    newlist = [] ##Create an empty new list
    for item in lst: ##Parse each sublist
        if item [1] == "District of Columbia": ##Check if the item in position 1 in the sublist equals District of Columbia
            newlist.append(item [0]) ##If so, add the item in position 0 in that sublist to empty new list created earlier.
        
    print(min(newlist)) ##Print the smallest value in the new list
            
firstDC(state_covid_data) ##Run the function! Spoiler alert: DC recorded its first case of COVID-19 on March 7, 2020.
```

```python
##Part 3: Write a function that takes in a state name as input (e.g. "Wisconsin") and outputs the date of its first case.
def first_case(state):
    """Finds earliest date when a given U.S. state recorded a case of COVID-19"""
    
    """This functions takes a state name as an argument, converts it into a string (not 100% sure if this is necessary
    but it does it anyway), creates a new empty list, checks if a sublist contains the requested state name, and if so, 
    appends the date in that sublist to the empty list. It then returns the minimum value in that new list, which is the
    earliest date that the requested state recorded a case of COVID-19."""
    state_string = str(state) ##Convert the state argument to a string. Included because some states have multiword names. 
    newlist = [] ##Create an empty new list
    for item in state_covid_data: ##Parse each sublist in the data set
        if item [1] == state_string: ##Check if the item in position 1 in the sublist equals the requested state
            newlist.append(item [0]) ##If so, add the item in position 0 in that sublist to empty new list created earlier.
    print(state_string, "recorded its first case of COVID-19 on", min(newlist)) ##Print the smallest value in the new list
            
first_case("Florida") ##Run the function! I tested this with random states, and it consistently worked.
```

```python
##Optional: Write a function that takes in a state name as input (e.g. "Wisconsin") 
##and outputs the date when the number of reported cases within the state exceeded 1000.
def thousand_cases(state):
    """Finds the earliest date when a given state had 1000 or more cases of COVID-19"""
    
    """This function takes a state name as an argument and converts it to a string. It then creates a new, empty list.
    It then evaluates each sublist in the data set, checks both if the sublist contains the requested state name AND if
    the case count is greater than or equal to 1000. If both of those conditions are true, it appends the date value from
    that sublist to the new empty list, and then prints the smallest value (i.e. the earliest date) in the new list."""
    state_string = str(state) ##Convert the state argument to a string. Included because some states have multiword names. 
    newlist = [] ##Create an empty new list
    for item in state_covid_data: ##Parse each sublist in the data set
        if (item [1] == state_string) and (1000 <= int(item[3])): ##Check if the item in position 1 in the sublist equals the requested state, and if there are 1000 or more cases
            newlist.append(item [0]) ##If so, add the item in position 0 in that sublist to empty new list created earlier.
    print(state_string, "reached 1,000 cases of COVID-19 on", min(newlist)) ##Print the smallest value in the new list
    
thousand_cases("Tennessee") ##Run the function! I also tested this with random states (and I compared the output to the raw data) and it consistently returned the correct dates.
```
