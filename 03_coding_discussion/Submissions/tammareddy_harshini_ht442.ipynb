### Task Initiation 

```python
# Importing Pandas tools 
import pandas as pd 
#Importing input data using pandas tool
df = pd.read_csv("chicago_summer_2018_crime_data.csv")
#printing input data imported from input csv file
df.head()
```

### Creating Output Dataframe

```python
# Filtering the required output columns from input data using groupby and creating converted_data for output data
#unstack function for count values
converted_data = df.groupby(['primary_type', 'day']).size().unstack(level=-1).reset_index().fillna(0)
days = list(df['day'].unique()) #unique days in column day
col_dict = {i: f'Day-{i}' for i in days}#Creating headers info - day
converted_data[days] = converted_data[days].apply(lambda x: round(x/x.sum(), 2), axis=1)#lamda function to calculate proportions
#renaming headers
converted_data = converted_data.rename(columns=col_dict).rename_axis(None, axis=1)
```

### Printing Output Data

```python
#printing the output dataframe
print(converted_data)
```
