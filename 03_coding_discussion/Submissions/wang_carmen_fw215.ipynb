```python
import pandas as pd 
```

```python
dat = pd.read_csv("chicago_summer_2018_crime_data.csv")
dat
```

```python
#get all primary type values
print(set(dat['primary_type']))
```

```python
#get all day values 
print(set(dat['day']))
```

```python
#convert the data frame to a new data frame which only includes primary_type, day, and count
new = pd.DataFrame(dat.filter(['primary_type','day']).groupby(['primary_type','day']).size().reset_index())

#rename the count column 
new.rename(columns = {'primary_type':'primary_type','day':'day',0:'count'}, inplace = True)

#create a new data frame using the pivot table
#individual count as the value, primary_type as index, day as columns
new_df = new.pivot_table(values = 'count', index = 'primary_type', columns = 'day',fill_value = 0)
#insert a new column for the sum of crimes
new_df['sum'] = new_df.sum(axis = 1)
#get the porportional data by (count/sum) 
#then drop the sum column 
new_df = new_df.div(new_df['sum'], 0).round(2).drop(['sum'], axis = 1)
```
