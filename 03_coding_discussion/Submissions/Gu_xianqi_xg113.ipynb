```python
import pandas as pd
```

```python
#read in data 
df = pd.read_csv("https://raw.githubusercontent.com/edunford/coding_discussions_ppol564_fall2020/master/03_coding_discussion/chicago_summer_2018_crime_data.csv")
df.head()
```

#### To get the number of crimes by the crime type and by each day

```python
df1 = (df
      .filter(['primary_type','day'])  #filter the dataframe by selecting crime type and day
      .groupby(['primary_type','day']) #group the dataframe by crime type and day
      .size())                         #count the number of observations

df1
```

#### Calculate the proportion of times that crime type was committed over all days of the month

```python
#get the number of observations by crime type 
primary_type = df.groupby('primary_type').size()
primary_type
```

```python
#calculate the proportion and round the value to second decimal place 
value = (df1/primary_type).round(2)
value
```

#### Convert the Panda Series to a dataframe in a wide format

```python
#convert the series to dataframe and rename the value column as Proportion
df_proportion = (value
                 .reset_index()
                 .rename(columns = {0:'Proportion'}))    
df_proportion
```

```python
#alter the structure from long to wide 
df_proportion.pivot_table(values = 'Proportion', columns = 'day', index = 'primary_type', fill_value = 0)
```
