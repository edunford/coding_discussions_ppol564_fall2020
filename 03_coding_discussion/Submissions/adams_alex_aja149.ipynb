```python
import pandas as pd
pd.set_option('display.max_rows', None) #This way the final data set won't by truncated and will show all rows and columns
pd.set_option('display.max_columns', None)
dta = pd.read_csv('C:/Users/alexi/coding_discussions_ppol564_fall2020/03_coding_discussion/chicago_summer_2018_crime_data.csv')
```

```python
df = dta[['month', 'day', 'primary_type']] #Copy the original data set but only include these relevant columns
```

```python
df = df.groupby(['primary_type','month','day']).size() #Group the data first by crime type, then month, then day
df = df.unstack(level=-1)                              #Unstack by day and use day as the columns
df = df.fillna(0)                                      #Replace all missing values with 0
df = df.div(df.sum(axis=1), axis=0)                    #Divide each element in the data frame by the sum of its row
                                                            #Sum of row = total number of crimes of that type in that month
df = df.round(decimals=2)                              #Round each element in the data frame to two decimal places
```
