```python
import pandas as pd
from dfply import *
```

```python
# reading the data
chig = pd.read_csv("chicago_summer_2018_crime_data.csv")

# Viewing the top 10 rows
(chig >>
head(10))
#df = pd.DataFrame(chig, )
```

It appears that every row is a crime.

```python
# Making a wide table
# Across the columns (on top): days
# Across the rows (down): primary-type
wideTable = (chig >>
    # group by primary_type for each day
    group_by(X.primary_type, X.day) >>
    # Spawn a new column called count that 
             # counts every crime per day
    # count = number of a specific crime that happened that one day
     summarize(count = n(X.primary_type))
).pivot_table(index = "primary_type", values = "count", columns = "day", fill_value = 0)

# Fill NAs with 0
#wideTable = wideTable.fillna(0)
# Sum of each crime per month
sumCrimePerMonth = (wideTable.sum(axis = 1))

wideTable["total"] = sumCrimePerMonth
# axis = 0: rows
# axis = 1: columns
# Because I am crossing all the columns in 1 row.
#
#print("Here is wide table")
#print(wideTable)

# for every column (name, not index) in wideTable
for column in wideTable.iloc[:, 0:31]:
    # but column refers to the name, not to an actual index
    # divide every value by the total
    # round every decimal to 2 places
    wideTable[column] = round(wideTable[column]/wideTable["total"], 2)
    
# Printing our new wideTable of proportions
wideTable
```
